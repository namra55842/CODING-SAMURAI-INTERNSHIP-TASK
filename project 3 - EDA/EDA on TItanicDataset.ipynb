```python
df['Parch'].value_counts()  # give the value of different atttributes in given column

#changing datatype
df['Survived'] =df['Survived'].astype('category')
df['Pclass'] =df['Pclass'].astype('category')
df['Sex'] =df['Sex'].astype('category')
df['Age'] =df['Age'].astype('category')
df['Embarked'] =df['Embarked'].astype('category')
df.info()
df.describe()
```

```python
# Convert Age back to numeric
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Fill missing values, then cast to int
df['Age'] = df['Age'].fillna(df['Age'].median()).astype('int64')

# Plot
sns.distplot(df['Age'])
print(df['Age'].skew())
print(df['Age'].kurt())
```

```python
sns.boxplot(df['Age'])
print("people btw 60 and 70 " , df[(df['Age']>60) & (df['Age']<70)].shape[0])
print("people btw 70 and 75 " , df[(df['Age']>=70) & (df['Age']<=75)].shape[0])
print("people above 75 " , df[(df['Age']>75)].shape[0])
```

```python
df['Fare'] = df['Fare'].astype('int64')
sns.distplot(df['Fare'])
print(df['Fare'].skew())
print(df['Fare'].kurt())
```

```python
# Feature engineering

df['familySize'] = df['Parch']+df['SibSp']
df.head()
```

```python
def family_type(num):
    if num==0:
        return"Alone"
    elif num>0 and num<=4:
        return "Medium"
    else:
        return "Large"

df['family_type'] = df['familySize'].apply(family_type)
df.head()
```

```python
# Finding quartiles
Q1 = np.percentile(df['Fare'], 25)   # 25th percentile
Q3 = np.percentile(df['Fare'], 75)   # 75th percentile

# IQR = Q3 - Q1
IQR = Q3 - Q1

# Lower and upper bounds for outliers
outlier_low = Q1 - 1.5 * IQR
outlier_high = Q3 + 1.5 * IQR

# Keep only values within the bounds
df = df[(df['Fare'] > outlier_low) & (df['Fare'] < outlier_high)]
```

```python
#one hot encoding
print(df.columns)
pd.get_dummies(data=df, columns=['Pclass', 'Sex', 'Embarked'], drop_first=True)
```
