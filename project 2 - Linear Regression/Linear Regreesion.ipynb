```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df.head(10)
```

```python
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'], df['per capita income (US$)'], color="red", marker="+")
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])    #fit means train the model usinf available data points       
```

```python
reg.predict([[2025]])
```

```python
reg.coef_         # value of m (slope)
```

```python
reg.intercept_   # value of b
```

```python
# y = m*x +b 
828.46507522*2025-1632210.7578554575
```

```python
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'], df['per capita income (US$)'], color="red", marker="+")
plt.plot(df['year'],reg.predict(df[['year']]),color='blue')
plt.show()
```

```python
d = pd.read_csv('year.csv')
d.head(10)
```

```python
p = reg.predict(d)
```
